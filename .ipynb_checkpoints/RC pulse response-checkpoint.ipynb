### Bode plot

Create the Bode plot of $$H(s)=\frac{1}{s+1}$$

```python
#create the transfer function
h = signal.lti([1], [1,1])
#set the range
w=np.logspace(-2,2)
#compute the magnitude and phase
w, mag, phase = signal.bode(h,w)
logplot(w,mag,phase,'1/(s+1)')
```
