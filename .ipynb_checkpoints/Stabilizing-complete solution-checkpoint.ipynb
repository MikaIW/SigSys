```python
%matplotlib inline
%run convenience.ipynb
from __future__ import print_function
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

np.set_printoptions(precision=2,suppress=True) # numpy output options

pi=np.pi
j=1j
```

### System stabilization

```python
sys = signal.lti([1], [1,-.01,1]) #original system without control
t,s=signal.step(sys,T=np.arange(0,300,.1))
timeplot(t,s,label='step response (V)')
```

```python
pzmap(sys)
```

```python
pzmap(sys,axis=[-.01,.01,-1.5,1.5])#rescaling
```

```python
kp=10 #proportional gain of 10
sys = signal.lti([kp], [1,-.01,1+kp]) # proportional control
pzmap(sys,axis=[-.01,.01,-4,4])#rescaling
```

```python
t,s=signal.step(sys,T=np.arange(0,300,.1))
timeplot(t,s,label='step response (V)')
```

```python
kp=-10 #proportional gain of -10
sys = signal.lti([kp], [1,-.01,1+kp]) # proportional control
pzmap(sys)
```

```python
t,s=signal.step(sys,T=np.arange(0,5,.1))
timeplot(t,s,label='step response (V)')
```

```python
ki=10 #integral gain of 10
sys = signal.lti([ki], [1,-.01,1,ki]) # integral control
pzmap(sys)
```

```python
t,s=signal.step(sys,T=np.arange(0,12,.1))
timeplot(t,s,label='step response (V)')
```
