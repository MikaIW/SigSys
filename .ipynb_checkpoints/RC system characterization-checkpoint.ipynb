```python
%matplotlib inline
%run convenience.ipynb
from __future__ import print_function
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

np.set_printoptions(precision=2,suppress=True) # numpy output options

pi=np.pi
j=1j
```

### Pulse response of an RC circuit

```python
#set the time constant
rc=7.
```

```python
#create the transfer function
sys = signal.lti([1/rc], [1,1/rc])
```

```python
#set the range
t=np.arange(0,50,.1)
```

```python
#create a unit area pulse
x = np.zeros(len(t))
pulse_width=2
pulse_start=10
x[np.logical_and(t>pulse_start,t<pulse_width+pulse_start)] = 1./pulse_width

plt.plot(t,x,lw=2)
myplotstyle()
plt.title('pulse input')
plt.ylabel('input (V)')
_=plt.xlabel('time (s)')
```

```python
t,state,y=signal.lsim(sys,x,t)
```

```python
plt.plot(t,x,lw=2)
plt.plot(t,y,lw=2)
myplotstyle()
plt.title('RC pulse response')
plt.ylabel('output (V)')
_=plt.xlabel('time (s)')
```

### Impulse response

```python
t,h=signal.impulse(sys,T=t)
plt.plot(t,h,lw=2)
myplotstyle()
plt.title('RC impulse response')
plt.ylabel('output (V)')
_=plt.xlabel('time (s)')
```
