### System characterization with Bode plot, step response and pole-zero map

```python
combinedplot(signal.lti([1,0],[1,1]),w=np.logspace(-2,2))
```

```python
combinedplot(signal.lti([1,0],[1,100,1]),w=np.logspace(-4,4))
```

```python
combinedplot(signal.lti([1,0],[1,1,1]),w=np.logspace(-2,2))
```

```python
combinedplot(signal.lti([1,0],[1,.1,1]),T=np.arange(0,120,.1),w=np.logspace(-2,2,200))
```

```python
combinedplot(signal.lti([1,-.01,1],[1,.01,1]),T=np.arange(0,1200,.5),axis=[-.05, .05, -1.5, 1.5])#Note the scale of the pole-zero map!
```

```python
t,s=signal.step(signal.lti([1,-.01,1],[1,.01,1]),T=np.arange(0,1200,.5),)
timeplot(t,s,label='step response')
_=plt.ylim([.98,1.02])#zoom in; note the amplitude!
```

```python
combinedplot(signal.lti([1,.01,1],[1,.011,1]),T=np.arange(0,1200,.5),axis=[-.05, .05, -1.5, 1.5])#Note the scale of the pole-zero map!
```

```python
t,s=signal.step(signal.lti([1,.01,1],[1,.011,1]),T=np.arange(0,1200,.5),)
timeplot(t,s,label='step response')
_=plt.ylim([.999,1.001])#zoom in; again, note the amplitude!
```
