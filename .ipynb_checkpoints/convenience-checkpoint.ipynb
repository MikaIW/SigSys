```python
'''
convenience functions
march 2105
'''

%matplotlib inline
from __future__ import print_function
import numpy as np
import matplotlib.pyplot as plt

# numpy output options
np.set_printoptions(precision=2,suppress=True)

pi=np.pi
j=1j
```

```python
def clean(x):
    tol=1e-10
    y=np.copy(x)
    y[abs(y) < tol] = 0.0
    return y
```

```python
def myplotstyle():
    'customize plot style'
    plt.margins(0.1)
    plt.axhline(color='k')
    plt.axvline(color='k')
    plt.grid(which='both')
```

```python
def complexplot(*args,**kwargs):
    '''
    plot complex sequences
    arguments: (range, values, style, title, xlabel)
    '''
    plot_range=args[0]
    values=args[1]
    style=args[2]
    title=args[3]
    xlabel_string=args[4]
    
    plt.figure(1,figsize=(16,4))#dpi=120 has no effect 
    plt.subplots_adjust(wspace=.15,hspace=0.03)
    plt.suptitle(title)
    
    plt.subplot(2,2,1)
    cleaned=clean(np.abs(values))
    plt.plot(plot_range,cleaned,style,**kwargs)
    plt.ylabel('magnitude')
    myplotstyle()
    
    plt.subplot(2,2,3)
    cleaned=clean(np.angle(values))
    cleaned[cleaned>pi-.01]=-pi#choose -pi over +pi
    plt.plot(plot_range,cleaned/pi,style,**kwargs)
    plt.ylabel('phase/$\pi$')
    myplotstyle()
    plt.xlabel(args[4])
    
    plt.subplot(2,2,2)
    cleaned=clean(np.real(values))
    plt.plot(plot_range,cleaned,style,**kwargs)
    plt.ylabel('real')
    myplotstyle()
    
    plt.subplot(2,2,4)
    cleaned=clean(np.imag(values))
    plt.plot(plot_range,cleaned,style,**kwargs)
    plt.ylabel('imaginary')
    myplotstyle()
    plt.xlabel(xlabel_string)
```

```python
def timeplot(*args,**kwargs):
    '''
    plot in time domain
    arguments: (range, values, title,ylabel)
    '''
    plot_range=args[0]
    values=args[1]
    title=args[2]
    ylabel_string=args[3]
    
    plt.figure(1,figsize=(16,4))#dpi=120 has no effect 
    plt.title(title)
    plt.plot(plot_range,values,lw=2)
    plt.ylabel(ylabel_string)
    myplotstyle()
    plt.xlabel('time (s)')
```
