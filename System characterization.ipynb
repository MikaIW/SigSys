```python
%matplotlib inline
%run convenience.ipynb
from __future__ import print_function
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

np.set_printoptions(precision=2,suppress=True) # numpy output options

pi=np.pi
j=1j
```

### System characterization

The system is described by the transfer function of series RC or series RLC circuit: $$H(s)=\frac{1/RC}{s+1/RC} \qquad \text{or} \qquad H(s)=\frac{1/LC}{s^2+(R/L)s+1/LC}$$

```python
C=1e-7
L=1e-2
R=50#400 or 50 ohm

#RLC
sys = signal.lti([1/(L*C)], [1,R/L,1/(L*C)])#create the transfer function
sys_string='RLC'#identifier for graphs

#RC
#sys = signal.lti([1/(R*C)], [1,1/(R*C)])#create the transfer function
#sys_string='RC'#identifier for graphs

time_scale=5*L/R#how long graphs should go for; scales inputs accordingly
t=np.linspace(0,10*L/R,500)#set the number of points OR
#t=np.arange(0,5*L/R,time_scale/500)#set the step size
```

```python
_t,h=signal.impulse(sys,T=t)
timeplot(t,h,label=sys_string+' impulse response (V)')
```

```python
_t,s=signal.step(sys,T=t)
timeplot(t,s,label=sys_string+' step response (V)')
```

```python
#create a delayed unit step
x = np.zeros(len(t))
pulse_start=t[-1]/5 #delay by a fifth of the end time of t_step
x[t>pulse_start] = 1

_t,y,state=signal.lsim(sys,x,t)#simulate the output, t is not the same as given in the input?

timeplot(t,x,label='time-shifted step (V)')
timeplot(t,y,label=sys_string+' shifted step response(V)')#BUG: plot dissapears if you add a space before (V)
```

```python
#create a unit area pulse
x = np.zeros(len(t))
pulse_width=.03*t[-1]#pulse width in fractions of t_step
pulse_start=t[-1]/5#delay by a fifth of the end time of t_step
x[np.logical_and(t>pulse_start,t<pulse_width+pulse_start)] = 1./pulse_width

_t,y,state=signal.lsim(sys,x,t)

timeplot(t,x,label='pulse (V)')
timeplot(t,y,label=sys_string+' pulse response (V)')
```

```python
timeplot(t,y,label=sys_string+' pulse response (V)')#plot only the pulse response in case their scale is very different
```

```python
x=np.zeros(len(t))
delay=t[-1]/5
x[t>delay]=10/t[-1]*np.sin(2*pi*2/t[-1]*(t[t>delay]-delay))#sin at a scaled inverse of the time scale starting at t=delay
timeplot(t,x,label='input: time-delayed sinusoid (V)')

_t,y,state=signal.lsim(sys,x,t)
timeplot(t,y,label=sys_string+' response: transient + sinusoidal steady-state (V)')

#overlay the impulse response to try to see the effect of the natural frequencies on the transient response
_t,h=signal.impulse(sys,T=t)
#shift by delay seconds and clip the last delay seconds from the plot
timeplot(t[t<(t[-1]-delay)]+delay,h[t<(t[-1]-delay)],label=sys_string+' impulse response (V)')
```

```python
import decimal
decimal.getcontext().prec=3
res_freq=decimal.Decimal(1/np.sqrt(L*C)*1/(2*pi))
print('resonant frequency =',res_freq.normalize().to_eng_string())

freqs=np.logspace(3,5,500)
_w, mag, phase = signal.bode(sys,w=2*pi*freqs)#Bode takes frequencies in rad/s and returns magnitudes in dB
logplot(freqs,10**(mag/20),phase,title=sys_string+' Bode plot',margins=0.4)
```

```python
#this should be exactly the same as the Bode plot
_w, Hs = signal.freqresp(sys,w=2*pi*freqs)
logplot(freqs,np.abs(Hs),np.angle(Hs)*180/pi,title=sys_string+' frequency response',margins=0.4) 
```
