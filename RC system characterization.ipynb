```python
%matplotlib inline
%run convenience.ipynb
from __future__ import print_function
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

np.set_printoptions(precision=2,suppress=True) # numpy output options

pi=np.pi
j=1j
```

```python
rc=5.#time constant
sys = signal.lti([1/rc], [1,1/rc])#create the transfer function
```

```python
t,h=signal.impulse(sys)
timeplot(t,h,'RC impulse response','output (V)')
```

```python
t,s=signal.step(sys)
timeplot(t,s,'RC step response','output (V)')
```

```python
w, mag, phase = signal.bode(sys)
logplot(w,mag,phase,'RC Bode plot')
```

```python
w, Hs = signal.freqresp(sys)
logplot(w,np.abs(Hs),np.angle(Hs)*180/pi,'RC frequency response')
```

```python
plt.plot(np.real(Hs),np.imag(Hs), 'b')
plt.plot(np.real(Hs), -np.imag(Hs), 'r')
myplotstyle()
```

```python
t=np.arange(0,50,.1)#set the range

#create a unit area pulse
x = np.zeros(len(t))
pulse_width=2
pulse_start=10
x[np.logical_and(t>pulse_start,t<pulse_width+pulse_start)] = 1./pulse_width

t,state,y=signal.lsim(sys,x,t)

timeplot(t,x,'','')
timeplot(t,y,'RC pulse response','output (V)')
timeplot(t,y,'RC pulse response','output (V)')
```
